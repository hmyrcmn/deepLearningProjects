```python
import cv2
import numpy as np

# Modeli yÃ¼kleyin ve konfigÃ¼rasyon dosyasÄ±nÄ± ayarlayÄ±n
model = cv2.dnn_DetectionModel("/content/Result4/runs/detect/train/weights/yolov8n.pt", "/content/Result4/runs/detect/train/weights/yolov8n.cfg")
model.setInputSize(640, 640)  # GiriÅŸ boyutunu ayarlayÄ±n
model.setInputScale(1.0 / 255)  # GiriÅŸ Ã¶lÃ§eÄŸini ayarlayÄ±n

# SÄ±nÄ±f adlarÄ±nÄ± ve renkleri ayarlayÄ±n
classes = ["without_mask", "with_mask", "mask_weared_incorrect", "without_hairnet", "with_hairnet", "hairnet_weared_incorrect"]
class_colors = [(0, 0, 255), (0, 255, 0)]  # Mavi ve YeÅŸil

# Video giriÅŸi alÄ±n
cap = cv2.VideoCapture("/content/hairnetinput3.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Nesne algÄ±lama yapÄ±n
    classes, scores, boxes = model.detect(frame, confThreshold=0.5, nmsThreshold=0.4)

    for (classId, score, box) in zip(classes, scores, boxes):
        label = classes[classId[0]]
        confidence = score[0]

        if "with_hairnet" in label and confidence > 0.5:
            color = class_colors[0]  # Mavi
        elif confidence > 0.5:
            color = class_colors[1]  # YeÅŸil
        else:
            continue

        (x, y, w, h) = box
        cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
        text = f"{label}: {confidence:.2f}"
        cv2.putText(frame, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    cv2.imshow("Video", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
